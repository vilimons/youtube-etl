```python
# Import of libraries

import pandas as pd
import os
```

```python
# Datasets names
list_of_names = ['BR_youtube_trending_data', 'CA_youtube_trending_data', 'DE_youtube_trending_data', 'FR_youtube_trending_data', 'GB_youtube_trending_data',\
                        'IN_youtube_trending_data', 'JP_youtube_trending_data', 'KR_youtube_trending_data', 'MX_youtube_trending_data', 'RU_youtube_trending_data',\
                            'US_youtube_trending_data']

# Empty list to receive datasets
dataframes_list = []

# appending datasets into the list
for i in range(len(list_of_names)):
    temp_df = pd.read_csv("./datasets/"+list_of_names[i]+".csv")
    dataframes_list.append(temp_df)
```

```python
# Verifying 
dataframes_list[1].head()
```

```python
# Drop of unuseful columns

for i, df in enumerate(dataframes_list):
    if 'Unnamed: 0' in dataframes_list[i].columns:
        dataframes_list[i].drop(columns=['Unnamed: 0'], inplace=True)
```

```python
dataframes_list[1].head()
```

```python
# Casting columns type from bool to string

for i, df in enumerate(dataframes_list):
    dataframes_list[i] = dataframes_list[i].astype({"comments_disabled": str, "ratings_disabled":str})
```

```python
# Verifying

dataframes_list[10].dtypes
```

```python
# Uploading files to required path

folderpath = "datasets/transformed_data/"  # output file type
for i, df in enumerate(dataframes_list, 0):
    filename = list_of_names[i] + "_transformed.parquet"
    filepath = os.path.join(folderpath, filename)
    df.to_parquet(filepath, index=False)
```
